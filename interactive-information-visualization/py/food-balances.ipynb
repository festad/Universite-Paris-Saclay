```python
# Locating the file
HOME = Path() / '../'
FAO = HOME / 'food-waste-dataset' / 'fao'
DATA = FAO / 'food-balances' / 'food-balances.csv'
df = pd.read_csv(DATA, encoding='latin-1')
```

```python
countries = ['France', 'Finland', 'India', 'Italy', 'China', 'United States of America', 'Japan']

gt_df = df.loc[(df['Area'].isin(countries))
             & (df['Item'] == 'Grand Total')
             & (df['Element'] == 'Food supply (kcal/capita/day)')]

print(gt_df.head())
print('-----------------')

gt_df = gt_df[['Area', 'Year', 'Value']]
gt_df = gt_df.pivot(index='Year', columns='Area', values='Value')

print(gt_df.head())
print('-----------------')

gt_df.plot(title='Food supply')
plt.xlabel('Year')
plt.ylabel('kcal/capita/day')
```
